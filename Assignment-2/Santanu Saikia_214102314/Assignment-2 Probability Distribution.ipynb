# Gaussian distribution

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
fig,axes=plt.subplots(nrows=1,ncols=3,figsize=(12,6))     #Generate three subplots
mu,sigma=0,1                                      #set mean and standard deviation
x=np.linspace(-6,6,2000)                          #Draw 2000 linearly spaced points between -6 to 6
y=1/(sigma *np.sqrt(2 * np.pi)) *np.exp( - (x -mu)**2 / (2 * sigma**2) )        #equation of PDF
axes[0].plot(x,y,lw=2, color='r',label='PDF')                      #Plot the  1st distribution
axes[0].set_xlabel('X')                                            #set x label
axes[0].set_ylabel('P(X)')                                         #set y label
axes[0].set_ylim([0,0.5])                                          #set limit of y-axis
axes[0].set_title('PDF \n mean=1,variance=0')                      #Give title 

#show mean,median and mode
axes[0].axvline(mu,ls='--',color='grey',label='mean\n median \n mode')
axes[0].legend(loc='best')                                         


mu_1,sigma_1=2,1                                                            #Set mean and standard deviation
x=np.linspace(-10,10,2000)
y=1/(sigma_1*np.sqrt(2 * np.pi)) *np.exp( - (x -mu_1)**2 / (2 * sigma_1**2) )             
axes[1].plot(x,y,lw=2, color='r',label='PDF')                               #Plot the second distribution                      
axes[1].set_xlim(-3,6)                                                      #set limit on x-axis
axes[1].set_ylim([0,0.5])                                                   #set limit on y-axis
axes[1].set_xlabel('X')                                                     #Give labelling
axes[1].set_ylabel('P(X)')
axes[1].set_title('PDF \n mean=2,variance=1')

#show mean,median and mode
axes[1].axvline(mu_1,ls='--',color='grey',label='mean\n median \n mode')
axes[1].legend(loc='best')

m,sig=4,2                                                                 #set mean and standard deviation
x=np.linspace(-12,12,2000)
y=1/(sig*np.sqrt(2 * np.pi)) *np.exp(-(x -m)**2 / (2 * sig**2) )                 
axes[2].plot(x,y,lw=2, color='r',label='PDF')                             #Plot the 3rd distribution
axes[2].set_xlim(-3,10)                                                   #set limit on x-axis
axes[2].set_ylim([0,0.3])                                                 #set limit on y-axis
axes[2].set_xlabel('X')                                                   #Give labelling
axes[2].set_ylabel('P(X)')
axes[2].set_title('PDF \n mean=4,variance=4') 

#show mean,median and mode
axes[2].axvline(m,ls='--',color='grey',label='mean\n median \n mode')
axes[2].legend(loc='best')

plt.tight_layout()

```

# Standard Normal

```python
import numpy as np
import matplotlib.pyplot as plt
mu, sigma= 0, 1                          #set mean and standard deviation
x=np.linspace(-6,6,1000)                 #Draw 1000 linearly spaced points between -6 to 6
y=1/(sigma*np.sqrt(2*np.pi))*np.exp(-(x-mu)**2/ (2*sigma**2))        #Equation of PDF 
plt.plot(x,y,'r',lw=2,label='PDF')                                               #plot the PDF

plt.vlines(mu,ymin=0.0,ymax=0.5,ls='--',color='grey',label='mean\n median \n mode')   #show mean,median,mode
plt.legend(loc='best')          
plt.xlabel('X')                                                      #Set X-label    
plt.ylabel('P(X)')                                                   #Set Y-label
plt.title('PDF')                                                     #Give title of the plot
```

# Exponential Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

l=0.5                                                     #set lambda value  
mu_1=1/l                                                  #mean of 1st subplot
var_1=1/(l)**2                                            #variance of 1st plot
fig,axes=plt.subplots(nrows=1,ncols=3,figsize=(10,7))    #Create 3 subplots
   
x=np.linspace(0,10,1000)                                #Generate 1000 linearly spaced points between 0 to 10 in x-axis
y=l*(np.exp(-l*x))                                     #Equation of Pdf
axes[0].plot(x,y,'r',lw=2)                             #Plot 1st Subplot
axes[0].set_xlabel('X')                                #Give x-label
axes[0].set_ylabel('P(X)')                             #Give y-label
axes[0].set_title('PDF Plot 1 \n lambda=0.5,mean=2 \n variance=4')          #Give title of the plot
axes[0].axvline(mu_1,ls='--',color='grey',label='mean')                     #plot vertical line to show mean
axes[0].axvline(var_1,ls='--',color='b',label='var')                        #plot vertical line to show variance
axes[0].legend()                

lam=0.4                                                #Set lambda value 
mu_2=1/lam
var_2=1/(lam)**2
x=np.linspace(0,12,1000)
y=lam*(np.exp(-lam*x))
axes[1].plot(x,y,'b',lw=2)                             #Plot 2nd subplot
axes[1].set_xlabel('X')                                
axes[1].set_ylabel('P(X)')
axes[1].set_xlim([0,12])
axes[1].set_title('PDF Plot 2 \n lambda=0.4,mean=2.5 \n variance=6.25')          
axes[1].axvline(mu_2,ls='--',color='grey',label='mean')
axes[1].axvline(var_2,ls='--',color='y',label='var')
axes[1].legend()           
lambd=0.2                                              #Set lambda value
mu_3=1/lambd
var_3=1/(lambd)**2
x=np.linspace(0,30,2000)
y=lambd*(np.exp(-lambd*x))
axes[2].plot(x,y,'y',lw=2)                             #Plot 3rd subplot
axes[2].set_xlabel('X')
axes[2].set_ylabel('P(X)')
axes[2].set_title('PDF Plot 3 \n lamda=0.2,mean=5 \n variance=25')              
axes[2].axvline(mu_3,ls='--',color='grey',label='mean')
axes[2].axvline(var_3,ls='--',color='r',label='var')
plt.tight_layout()                                    #Prevent overlapping of subplots
```

# Chi-Squared Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

#x-axis ranges from 0 to 20 with .001 steps

x = np.arange(0, 20, 0.001)  
fig,axes=plt.subplots(nrows=1,ncols=3,figsize=(10,6))        #create 3 subplots
#plot Chi-square distribution with 2 degrees of freedom

axes[0].plot(x, chi2.pdf(x, df=2),'r',label='df:2')
axes[0].set_xlabel('X')
axes[0].set_ylabel('F(X)')
axes[0].set_title('PDF plot 1 \n d.o.f.=2')                
df = 2
mean, var= chi2.stats(df, moments='mv')         
print('Mean of 1st plot:',mean)
print('Variance of 1st plot:',var)

#plot Chi-square distribution with 4 degrees of freedom
axes[1].plot(x,chi2.pdf(x,df=4),label='df:4')              
axes[1].set_xlabel('X')
axes[1].set_ylabel('F(X)')
axes[1].set_title('PDF plot 2\n d.o.f.=4')
df = 4
mean_1, var_1 = chi2.stats(df, moments='mv')    #calculate mean and variance of 2nd subplot
print('Mean of 2nd plot:',mean_1)
print('Variance of 2nd plot:',var_1)

#plot Chi-square distribution with 8 degrees of freedom
axes[2].plot(x, chi2.pdf(x, df=8),'black',label='df:8')   
axes[2].set_xlabel('X')
axes[2].set_ylabel('F(X)')
axes[2].set_title('PDF plot 3 \n d.o.f.=8')
df = 8
mean_2, var_2 = chi2.stats(df, moments='mv')    #calculate mean and variance of 3rd subplot
print('Mean of 3rd plot:',mean_2)
print('Variance of 3rd plot:',var_2)
axes[0].legend()
axes[1].legend()
axes[2].legend()
plt.tight_layout()
```

# Discrete Distribution

# (a)

 The Probability mass function follows a Bernoulli distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
fig,axes=plt.subplots(nrows=1,ncols=3,figsize=(10,6))
p=0.1
Data={0:p, 1:1-p}
m=1-p
var=p*(1-p)
print('Mean of First plot:',m)
print('Variance of First plot:',var)
for x,y in Data.items():
    axes[0].plot((x,x),(0,y),'r',lw=2,marker='o')
    axes[0].set_xlabel('X')
    axes[0].set_ylabel('P(X)')
    axes[0].set_title('PMF \n p=0.1')

p_1=0.5
Data_1={0:0.5,1:0.5}
m1=1-p_1
var1=p_1*(1-p_1)
print('Mean of Second plot:',m1)
print('Variance of Second plot:',var1)
for x,y in Data_1.items():
    axes[1].plot((x,x),(0,y),'r',lw=2,marker='o')
    axes[1].set_xlabel('X')
    axes[1].set_ylabel('P(X)')
    axes[1].set_title('PMF \n p=0.5')

p_2=0.8
Data_2={0:p, 1:1-p}
m2=1-p_2
var2=p_2*(1-p_2)
print('Mean of Third plot:',m2)
print('Variance of Third plot:',var2)
for x,y in Data_2.items():
    axes[2].plot((x,x),(0,y),'r',lw=2,marker='o')
    axes[2].set_xlabel('X')
    axes[2].set_ylabel('P(X)')
    axes[2].set_title('PMF \n p=0.2')
plt.tight_layout()
```

# (c) 

The distribution is known as Geometric Distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

Data={1:0.5, 2:0.5*0.5, 3:0.5*(0.5)**2, 4:0.5*(0.5)**3, 5:0.5*(0.5)**4, 6:0.5*(0.5)**5, 7:0.5*(0.5)**6, 8:0.5*(0.5)**7, 
      9:0.5*(0.5)**8, 10:0.5*(0.5)**9,11:0.5*(0.5)**10}
p=0.5
m=(1-p)/p
var=(1-p)/p**2
print('Mean of plot:',m)
print('Variance of plot:',var)
for x,y in Data.items():
    plt.plot((x,x),(0,y),'r',lw=2,marker='o')
    plt.xlabel('X')
    plt.ylabel('P(X)')
    plt.title('PMF \n p=0.5')
```

# (d)

The Distribution Followed by X is Poisson Distribution.When the number of trials are finite,the distribution becomes Binomial 
which is in case of Second distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import poisson

mu = 5                                          #set mean
mean, var = poisson.stats(mu, moments='mv')     #calculate mean and variance
print('Mean:',mu)
print('Variance:',var)
x= np.arange(poisson.ppf(0.001, mu),poisson.ppf(0.995, mu))            #set points on x-axis for which probibilties are between
                                                                       #0.001 and 0.995 on y-axis
plt.plot(x, poisson.pmf(x, mu), 'bo', ms=5, label='poisson pmf')       #plot the PMF
plt.vlines(x, 0, poisson.pmf(x, mu), colors='b', lw=2, alpha=0.5)      
plt.xlabel('X')                                                        #Give labellig
plt.ylabel('P(X)')
plt.title('PMF')
```
