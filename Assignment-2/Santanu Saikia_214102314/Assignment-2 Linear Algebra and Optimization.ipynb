```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import numpy.linalg as la

A = np.array([[-1,2],[3,4]])    
print('A:\n',A)
results=la.eig(A)                                #calculate the eigen values and vectors of matrix A
print('eigen values A:',results[0])              #print eigen values
print('eigen vectors A:\n',results[1])           #print eign vectors
print('\n A: \n',A)
D=A.T                                          #for checking the validity of covariance matrix
print('Transpose of A: \n',D)
print('A is not a valid Co-variance Matrix ')

B=np.array([[2,4],[6,8]])
print('\n B:\n',B)
results_1=la.eig(B)
print('eigen values B:',results_1[0])
print('eigen vectors B:\n',results_1[1])
print('\n B: \n',B)
E=B.T
print('Transpose of B: \n',E)
print('B is not a valid Co-variance Matrix ')

C=np.array([[2,5],[5,7]])
print('\nC:\n',C)
results_2=la.eig(C)
print('\n eigen values C:',results_2[0])
print('eigen vectors C:\n',results_2[1])
print('C: \n',C)
print("\n")
F=C.T
print('Transpose of C: \n',F)
print('eigen values of C are not positive.')
print('C is not a valid Co-variance Matrix ')
```

```python
def f(x1,x2,mat):                      #For calculating the value of x(transpose)Ax
    x = np.array([x1,x2])
    x = x.reshape((2,1))
    return np.dot(np.dot(x.T,mat),x)
```

```python
import numpy as np
import numpy.linalg as la
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
%matplotlib inline

A = np.array([[-1,2],[3,4]])
print('A:\n',A)
results=la.eig(A)
print('eigen values A:',results[0])
x1=np.arange(-20,20,0.4)
x2=np.arange(-20,20,0.4)
x11=[]
x21=[]
y1=[]
for i in x1:
    for j in x2:
        x11.append(i)
        x21.append(j)
        y1.append(f(i,j,A)[0][0])

y1=np.array(y1)
x11=np.array(x11)
x21=np.array(x21)
ax = plt.axes(projection='3d')
ax.plot3D(x11,x21,y1, 'b')
ax.plot(x11,x21,y1, 'b')
plt.title("plot of 1st quadratic equation")
plt.xlabel("x1")
plt.ylabel("x2")  
plt.show()


```

# 4.D=np.array([[9,0], [0,1]]), gradient= 18*x1 i+ 2*x2 j, stationary point(x1,x2)=(0,0)
# 5.E=np.array([[1,0], [0,9]]), gradient= 2*x1 i+ 18*x2 j, stationary point(x1,x2)=(0,0)
# 6.F=np.array([[9,5], [5,4]]), gradient= (18*x1 + 10*x2)i + (8*x2 + 10*x1)j, stationary point(x1,x2)=(0,0)
# 7.H=np.array([[1,-1], [-1,1]]), gradient= (2*x1-2*x2)i + (2*x2 - 2*x1)j, stationary point: x1=x2
